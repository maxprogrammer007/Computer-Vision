```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
```

```python
# defining the imshow function
def imshow(title = 'Image',image = None,size = 10):
    w,h = image.shape[1],image.shape[0]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
# Arithmetic ops on images - These are simple ops which allows us to directly add / sub to the color intensities of the image
# calculates the per element sum of the two images

img = cv2.imread('C:\\Users\\abhin\\OneDrive\\Documents\\GitHub\\Computer-Vision\\data\\HappyFish.jpg')
imshow('Original Image',img)
```

```python
# create a matrix of ones then multiply it by a scalar of 100
# this gives us a matrix with same dimensions as the image but with all values as 100

M = np.ones(img.shape,dtype = 'uint8')*100
print(M)
```

```python
# converting into gray scale
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
imshow('Gray Image',gray)

M = np.ones(gray.shape,dtype = 'uint8')*100
```

```python
# iincreasing brightness

added = cv2.add(gray,M)
imshow('Added Image',added)
```

```python
added2 = gray + M
imshow('Added Image 2',added2)
```

```python
# Decreasing brightness 

subtracted = cv2.subtract(gray,M)
imshow('Subtracted Image',subtracted)

subtracted2 = gray - M
imshow('Subtracted Image 2',subtracted2)
```
