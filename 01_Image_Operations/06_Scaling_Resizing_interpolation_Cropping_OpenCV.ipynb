```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Resizing the image 
# cv2.resize(image, dsize(output image size), x scale, y scale, interpolation)
# if dsize is none then x and y scale should be provided
# list of interpolation methods - cv2.INTER_AREA, cv2.INTER_CUBIC, cv2.INTER_LINEAR, cv2.INTER_NEAREST  , cv2.INTER_LANCZOS4

# cv.INTER_AREA - resampling using pixel area relation. It may be a preferred method for image decimation, as it gives moire-free results. But when the image is zoomed, it is similar to the INTER_NEAREST method.
# cv.INTER_CUBIC - a bicubic interpolation over 4x4 pixel neighborhood
# cv.INTER_LINEAR - a bilinear interpolation over 2x2 pixel neighborhood
# cv.INTER_LANCZOS4 - a Lanczos interpolation over 8x8 pixel neighborhood
# cv.INTER_NEAREST - a nearest-neighbor interpolation

```

```python
# defining our imshow function

def imshow(title = 'Image',image = None,size = 10):
    w,h = image.shape[1],image.shape[0]
    aspect_ratio = w/h
    plt.figure(figsize=(size*aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
 # reading the image
img = cv2.imread('C:\\Users\\abhin\\OneDrive\\Documents\\GitHub\\Computer-Vision\\data\\aero3.jpg') 
imshow('Original Image',img)
```

```python
# if no interpolation is defined then it uses cv2.INTER_LINEAR by default 
# lets make our image 1/4th of its original size

image_scaled = cv2.resize(img,None,fx = 0.75,fy = 0.75)
imshow('Scaled Image',image_scaled)
```

```python
# lets make our image 2 times of its original size using cv2.INTER_CUBIC interpolation
image_scaled2 = cv2.resize(img,None,fx = 2,fy = 2, interpolation= cv2.INTER_CUBIC)
imshow('Scaled Image',image_scaled2)
```
